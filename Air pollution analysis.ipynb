```python
plt.figure(figsize=(10, 5))
plt.hist(data['CO'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('CO')
plt.ylabel('TOTAL_CONCENTRATION')
plt.title('CONCENTARIONS OF EACH GAS')
plt.show()
```

```python
gender_counts = data['Gender'].value_counts()

# using Matplotlib
plt.figure(figsize=(8, 6))
plt.bar(gender_counts.index, gender_counts.values, color=['blue', 'pink','black'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Number of Males and Females')
plt.show()

```

```python
# Alternatively, using Seaborn
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=data, palette=['black','blue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Number of Males and Females')
plt.show()
```

```python
country = data['ParentLocation'].value_counts()

# Plot the counts using Matplotlib
plt.figure(figsize=(20, 10))
plt.bar(country.index, country.values, color=['blue', 'pink','black'])
plt.xlabel('country')
plt.ylabel('Count')
plt.title('Number of countries')
plt.show()
```

```python
total_concentration = data[['CO', 'C6H6', 'NMHC', 'SO4', 'NOX']].sum()
```

```python
plt.figure(figsize=(5, 5))
plt.pie(total_concentration, labels=total_concentration.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of gases')
plt.show()
```
